# Lab1: Data Visualization - Iris Species

## 1 About Dataset

![](image/Iris.png)

Dataset Source: https://www.kaggle.com/datasets/uciml/iris


### 1.1 Background

The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

### 1.2 Content

The columns in this dataset are:

- Id
- SepalLengthCm
- SepalWidthCm
- PetalLengthCm
- PetalWidthCm
- Species


## 2 Data Preprocessing
Set up coding environment

```python
import numpy as np
import pandas as pd
from pandas.plotting import parallel_coordinates
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_palette("Set2")
sns.set_style("darkgrid")
current_palette = sns.color_palette()
print("Setup Complete")
```

Load and examine the data

```python
df = pd.read_csv('data/Iris.csv')
df.head()
```

```python
df["Species"].value_counts()
```

We can find that the dataset contains three iris species, namely Setosa, Versicolor and Virginica.

## 3 Data Visualization

### 3.1 Density chart

Density plots allow to visualize the distribution of a numeric variable for one or several groups.

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

sns.kdeplot(df.where(df.Species == "Iris-setosa")["SepalLengthCm"], color=current_palette[0], shade=True, ax=axs[0, 0])
sns.kdeplot(df.where(df.Species == "Iris-setosa")["SepalWidthCm"], color=current_palette[1], shade=True, ax=axs[0, 1])
sns.kdeplot(df.where(df.Species == "Iris-setosa")["PetalLengthCm"], color=current_palette[2], shade=True, ax=axs[1, 0])
sns.kdeplot(df.where(df.Species == "Iris-setosa")["PetalWidthCm"], color=current_palette[3], shade=True, ax=axs[1, 1])
```

From the charts above, we can draw the following conclusions:

- For Setosa, the maximum probability density of sepal length and width is separately 5.0cm and 3.3cm.
-  For Setosa, the maximum probability density of petal length and width is separately 1.5cm and 0.2cm.

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

sns.kdeplot(df.where(df.Species == "Iris-versicolor")["SepalLengthCm"], color=current_palette[0], shade=True, ax=axs[0, 0])
sns.kdeplot(df.where(df.Species == "Iris-versicolor")["SepalWidthCm"], color=current_palette[1], shade=True, ax=axs[0, 1])
sns.kdeplot(df.where(df.Species == "Iris-versicolor")["PetalLengthCm"], color=current_palette[2], shade=True, ax=axs[1, 0])
sns.kdeplot(df.where(df.Species == "Iris-versicolor")["PetalWidthCm"], color=current_palette[3], shade=True, ax=axs[1, 1])
```

From the charts above, we can draw the following conclusions:

- For Versicolor, the maximum probability density of sepal length and width is separately 5.7cm and 2.9cm.
-  For Versicolor, the maximum probability density of petal length and width is separately 4.5cm and 1.3cm.

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

sns.kdeplot(df.where(df.Species == "Iris-virginica")["SepalLengthCm"], color=current_palette[0], shade=True, ax=axs[0, 0])
sns.kdeplot(df.where(df.Species == "Iris-virginica")["SepalWidthCm"], color=current_palette[1], shade=True, ax=axs[0, 1])
sns.kdeplot(df.where(df.Species == "Iris-virginica")["PetalLengthCm"], color=current_palette[2], shade=True, ax=axs[1, 0])
sns.kdeplot(df.where(df.Species == "Iris-virginica")["PetalWidthCm"], color=current_palette[3], shade=True, ax=axs[1, 1])
```

From the charts above, we can draw the following conclusions:

- For Virginica, the maximum probability density of sepal length and width is separately 6.3cm and 3.0cm.
-  For Virginica, the maximum probability density of petal length and width is separately 5.5cm and 1.9cm.

### 3.5 Scatter plot

A scatter plot displays the relationship between 2 numeric variables. Each data point is represented as a circle.

```python
fig, axs = plt.subplots(2, 3, figsize=(14, 6))

sns.regplot(x=df.where(df.Species == "Iris-setosa")["SepalLengthCm"], y=df.where(df.Species == "Iris-setosa")["SepalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5}, color=current_palette[0], ax=axs[0, 0])
sns.regplot(x=df.where(df.Species == "Iris-versicolor")["SepalLengthCm"], y=df.where(df.Species == "Iris-versicolor")["SepalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5}, color=current_palette[1], ax=axs[0, 1])
sns.regplot(x=df.where(df.Species == "Iris-virginica")["SepalLengthCm"], y=df.where(df.Species == "Iris-virginica")["SepalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5}, color=current_palette[2], ax=axs[0, 2])

sns.regplot(x=df.where(df.Species == "Iris-setosa")["PetalLengthCm"], y=df.where(df.Species == "Iris-setosa")["PetalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5},color=current_palette[0], ax=axs[1, 0])
sns.regplot(x=df.where(df.Species == "Iris-versicolor")["PetalLengthCm"], y=df.where(df.Species == "Iris-versicolor")["PetalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5}, color=current_palette[1], ax=axs[1, 1])
sns.regplot(x=df.where(df.Species == "Iris-virginica")["PetalLengthCm"], y=df.where(df.Species == "Iris-virginica")["PetalWidthCm"], line_kws={"color":"r","alpha":0.7,"lw":5}, color=current_palette[2], ax=axs[1, 2])
```

From the charts above, we can draw the following conclusions:

- For all three species, sepal length is positively correlated with sepal width and petal length is also positively correlated with petal width.


### 3.6 Correlogram

A correlogram or correlation matrix allows to analyse the relationship between each pair of numeric variables of a matrix. The correlation is visualised as a scatterplot. The diagonal represents the distribution of each variable with a histogram or a density plot.

```python
sns.pairplot(df.drop("Id", axis=1), kind="scatter", hue="Species", markers=["o", "s", "D"])
```

From the charts above, we can draw the following conclusions:

- Setosa is linearly separable from the other two, but the other two are not linearly separable from each other.

### 3.7 Contour plot

A contour plot allows to visualize the distribution of a combination of 2 numeric variables. It always have a variable represented on the X axis, the other on the Y axis. Then the number of observations within a particular area of the 2D space is counted and represented with a color gradient.

```python
fig, axs = plt.subplots(2, 3, figsize=(14, 6))

sns.kdeplot(x=df.where(df.Species == "Iris-setosa")["SepalLengthCm"], y=df.where(df.Species == "Iris-setosa")["SepalWidthCm"], cmap="Greens", shade=True, bw_adjust=.5, ax=axs[0, 0])
sns.kdeplot(x=df.where(df.Species == "Iris-versicolor")["SepalLengthCm"], y=df.where(df.Species == "Iris-versicolor")["SepalWidthCm"], cmap="Reds", shade=True, bw_adjust=.5, ax=axs[0, 1])
sns.kdeplot(x=df.where(df.Species == "Iris-virginica")["SepalLengthCm"], y=df.where(df.Species == "Iris-virginica")["SepalWidthCm"], cmap="Blues", shade=True, bw_adjust=.5, ax=axs[0, 2])

sns.kdeplot(x=df.where(df.Species == "Iris-setosa")["PetalLengthCm"], y=df.where(df.Species == "Iris-setosa")["PetalWidthCm"], cmap="Greens", shade=True, bw_adjust=.5, ax=axs[1, 0])
sns.kdeplot(x=df.where(df.Species == "Iris-versicolor")["PetalLengthCm"], y=df.where(df.Species == "Iris-versicolor")["PetalWidthCm"], cmap="Reds", shade=True, bw_adjust=.5, ax=axs[1, 1])
sns.kdeplot(x=df.where(df.Species == "Iris-virginica")["PetalLengthCm"], y=df.where(df.Species == "Iris-virginica")["PetalWidthCm"], cmap="Blues", shade=True, bw_adjust=.5, ax=axs[1, 2])
```

From the charts above, we can draw the following conclusions:

- For Setosa, the most common individual has a sepal length of 5.1cm and a sepal width of 5.0cm, a petal length of 1.4cm and a petal width of 0.2cm.
- For Versicolor, the most common individual has a sepal length of 5.8cm and a sepal width of 5.7cm, a petal length of 4.1cm and a petal width of 1.3cm.
- For Virginica, the most common individual has a sepal length of 6.3cm and a sepal width of 3.0cm, a petal length of 5.1cm and a petal width of 1.9cm.
